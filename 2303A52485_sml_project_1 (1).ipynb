# Linear-Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train, y_train)
y_pred = model.predict(x_test)
print(y_pred)
```

```python
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
rmse = mean_squared_error(y_test, y_pred, squared=False)
mse = mean_squared_error(y_test, y_pred)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print("Root Mean Squared Error:", rmse)
print("Mean Squared Error:", mse)
print("R-squared:", r2)
print("Mean Absolute Error:", mae)
```


# GRAPH PLOTING






```python
import matplotlib.pyplot as plt

def plot_histograms(mse, rmse, mae, model_name):
    plt.figure(figsize=(12, 8))
        # MSE Histogram
    plt.hist(mse, bins=5, color='red', alpha=0.7, label='MSE')
    plt.title(f'{model_name} - MSE')
    plt.xlabel('MSE')
    plt.ylabel('Frequency')
    plt.legend()
    plt.show()

      # RMSE Histogram
    plt.hist(rmse, bins=5, color='blue', alpha=0.7, label='RMSE')
    plt.title(f'{model_name} - RMSE')
    plt.xlabel('RMSE')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid()
    plt.show()

       # MAE Histogram
    plt.hist(mae, bins=5, color='green', alpha=0.7, label='MAE')
    plt.title(f'{model_name} - MAE')
    plt.xlabel('MAE')
    plt.ylabel('Frequency')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
plot_histograms(lr_boot_mse, lr_boot_rmse, lr_boot_mae, 'Linear Regression')
```

```python
plot_histograms(lasso_boot_mse, lasso_boot_rmse, lasso_boot_mae, 'Lasso Regression')
```
