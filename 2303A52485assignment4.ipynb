1.Read the data with pandas and find features and target variables

2.Normalize the data with min-max scaling

3.Split the data into train and test.

```python
import pandas as pd
df = pd.read_csv("/content/train.csv")
#print(df)
#1
target = df.columns[-1]
y=df[target]
print(y)
cf =df.drop('price_range',axis=1) # Use 'df' instead of 'data'
print(cf)

#2
bf = cf-cf.min()/(cf.max()-cf.min())
print(bf)
#3
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(bf,y,test_size=0.2) # Use column names 'cf' to select features
print(x_train)
print(x_test)
print(y_train)
print(y_test)
```
