```python
import pandas as pd
b = pd.read_csv("/content/diabetes_data_upload.csv")
print(b)

columns= ['Age','Gender','polyuria','Polydipsia','sudden weight loss','weakness','Polyphagia','Genital thrush','visual blurring','Itching','Irritability','delayed healing','partial paresis','muscle stiffness','Alopecia','Obesity','class']
b.columns = columns

print(b.dtypes)
b = b.replace('No',0)
b = b.replace('Yes',1)
b = b.replace('Male',1)
b = b.replace('Female',0)
b = b.replace('Positive',1)
b = b.replace('Negative',0)
print(b)
```

```python
x = b['class']
print(b)
y= b.drop('class',axis=1)
print(y)
```
